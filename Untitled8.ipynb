```python
class Persona:
    def __init__(self, nombre, edad, genero):
        self.nombre = nombre
        self.edad = edad
        self.genero = genero

class Paciente(Persona):
    total_pacientes = 0

    def __init__(self, nombre, edad, genero, numero_historia_clinica):
        super().__init__(nombre, edad, genero)
        self.numero_historia_clinica = numero_historia_clinica
        Paciente.total_pacientes += 1

class Doctor(Persona):
    def __init__(self, nombre, edad, genero, especialidad):
        super().__init__(nombre, edad, genero)
        self.especialidad = especialidad

class Hospital:
    def __init__(self, nombre, ubicacion, limite_pacientes):
        self.nombre = nombre
        self.ubicacion = ubicacion
        self.pacientes = []
        self.limite_pacientes = limite_pacientes

    def registrar_paciente(self, paciente):
        if len(self.pacientes) >= self.limite_pacientes:
            print("El hospital está saturado. No se puede agregar más pacientes.")
        else:
            # Verificar si el paciente ya está registrado
            for p in self.pacientes:
                if p.numero_historia_clinica == paciente.numero_historia_clinica:
                    print(f"El paciente con número de historia clínica {paciente.numero_historia_clinica} ya está registrado.")
                    return
            self.pacientes.append(paciente)
            print(f"Paciente {paciente.nombre} registrado con éxito.")

    def total_pacientes_registrados(self):
        return len(self.pacientes)

class ConsultaMedica:
    def __init__(self, doctor, paciente, fecha):
        self.doctor = doctor
        self.paciente = paciente
        self.fecha = fecha

    def mostrar_informacion(self):
        print("Consulta Médica:")
        print("Fecha:", self.fecha)
        print("Doctor:")
        print("Nombre:", self.doctor.nombre)
        print("Edad:", self.doctor.edad)
        print("Género:", self.doctor.genero)
        print("Especialidad:", self.doctor.especialidad)
        print("Paciente:")
        print("Nombre:", self.paciente.nombre)
        print("Edad:", self.paciente.edad)
        print("Género:", self.paciente.genero)
        print("Número de Historia Clínica:", self.paciente.numero_historia_clinica)

# Ejemplo de uso
if __name__ == "__main__":
    hospital = Hospital("Hospital XYZ", "Ciudad ABC", 3)  # Establecer un límite de 2 pacientes

    paciente1 = Paciente("Paciente1", 30, "Masculino", "HIS001")
    paciente2 = Paciente("Paciente2", 25, "Femenino", "HIS002")
    paciente3 = Paciente("Paciente3", 40, "Femenino", "HIS003")  # Intentar agregar un tercer paciente
    doctor1 = Doctor("Doctor1", 40, "Masculino", "Cardiología")
    paciente4 = Paciente("Paciente4", 40, "Femenino", "HIS004")  # Intentar agregar un tercer paciente

    hospital.registrar_paciente(paciente1)  # Debe registrarse correctamente
    hospital.registrar_paciente(paciente2)  
    hospital.registrar_paciente(paciente3)  
    hospital.registrar_paciente(paciente4)

    consulta1 = ConsultaMedica(doctor1, paciente1, "2023-08-30")
    consulta1.mostrar_informacion()

    print("Total de pacientes registrados en el hospital:", hospital.total_pacientes_registrados())
```
