```python
from PyQt5 import QtWidgets, uic, QtGui
from pydicom import dcmread
import numpy as np
import cv2
import sys
import glob

class MainWindow(QtWidgets.QMainWindow):
    def __init__(self):
        super(MainWindow, self).__init__()
        uic.loadUi('VentanaPrincipal.ui', self)
        self.pushButton.clicked.connect(self.on_pushButton_clicked)  # Iniciar sesion
        self.pushButton_2.clicked.connect(self.on_pushButton_2_clicked)  # Salir

    def on_pushButton_clicked(self):
        # This will open a new window when the 'Iniciar sesion' button is clicked
        self.window = Ventana1()
        self.window.show()

    def on_pushButton_2_clicked(self):
        # This will close the application when the 'Salir' button is clicked
        QtWidgets.QApplication.quit()

class Ventana1(QtWidgets.QMainWindow):
    def __init__(self):
        super(Ventana1, self).__init__()
        uic.loadUi('Ventana1.ui', self)
        self.pushButton.clicked.connect(self.on_pushButton_clicked)  # Ver tomografia
        self.pushButton_2.clicked.connect(self.on_pushButton_2_clicked)  # Datos interesantes de anatomia
        self.pushButton_3.clicked.connect(self.on_pushButton_3_clicked)  # Anterior

    def on_pushButton_clicked(self):
        # This will open a new window when the 'Ver tomografia' button is clicked
        self.window = Ventana2()
        self.window.show()

    def on_pushButton_2_clicked(self):
        # This will open a new window when the 'Datos interesantes de anatomia' button is clicked
        self.window = Ventana3()
        self.window.show()

    def on_pushButton_3_clicked(self):
        # This will open the main window when the 'Anterior' button is clicked
        self.window = MainWindow()
        self.window.show()

class Ventana2(QtWidgets.QMainWindow):
    def __init__(self):
        super(Ventana2, self).__init__()
        uic.loadUi('Ventana2.ui', self)
        self.pushButton.clicked.connect(self.on_pushButton_clicked)  # Anterior
        self.display_images()

    def display_images(self):
        # Add your logic here to display .dcm images using OpenCV
        image_folder = 'imagenes'  # replace with your actual folder path
        for filename in glob.glob(image_folder + '/*.dcm'):
            ds = dcmread(filename)
            image = np.array(ds.pixel_array)
            image = cv2.normalize(image, None, 0, 255, cv2.NORM_MINMAX, dtype=cv2.CV_8U)
            qimage = QtGui.QImage(image, image.shape[1], image.shape[0], image.strides[0], QtGui.QImage.Format_Grayscale8)
            self.label_2.setPixmap(QtGui.QPixmap.fromImage(qimage))

    def on_pushButton_clicked(self):
        # This will open the main window when the 'Anterior' button is clicked
        self.window = Ventana1()
        self.window.show()

class Ventana3(QtWidgets.QMainWindow):
    def __init__(self):
        super(Ventana3, self).__init__()
        uic.loadUi('Ventana3.ui', self)
        self.pushButton.clicked.connect(self.on_pushButton_clicked)  # Anterior
        self.pushButton_2.clicked.connect(self.on_pushButton_2_clicked)  # Ver respuesta

    def on_pushButton_clicked(self):
        # This will open the main window when the 'Anterior' button is clicked
        self.window = Ventana1()
        self.window.show()

    def on_pushButton_2_clicked(self):
        # This will open a new window when the 'Ver respuesta' button is clicked
        msg = QtWidgets.QMessageBox()
        msg.setWindowTitle("Respuesta")
        msg.setText("La respuesta correcta es '206 huesos'")
        msg.exec_()

app = QtWidgets.QApplication(sys.argv)
window = MainWindow()
window.show()
sys.exit(app.exec())
```
