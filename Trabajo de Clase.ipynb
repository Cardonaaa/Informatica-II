```python
class Persona:
    def __init__(self, nombre, cedula, genero):
        self._nombre = nombre
        self._cedula = cedula
        self.genero = genero
    
    def ver_cedula(self):
        return self._cedula

class PersonalMedico(Persona):
    def __init__(self, nombre, cedula, genero, turno):
        super().__init__(nombre, cedula, genero)
        self.turno = turno

    def ver_cedula(self):
        return self._cedula

class Enfermero(PersonalMedico):
    def __init__(self, nombre, cedula, genero, turno, rango):
        super().__init__(nombre, cedula, genero, turno)
        self._rango = rango

class Medico(PersonalMedico):
    def __init__(self, nombre, cedula, genero, turno, especialidad):
        super().__init__(nombre, cedula, genero, turno)
        self._especialidad = especialidad

class Paciente(Persona):
    def __init__(self, nombre, cedula, genero, servicio):
        super().__init__(nombre, cedula, genero)
        self.servicio = servicio

# Lista para almacenar pacientes, enfermeros y médicos
pacientes = []
enfermeros = []
medicos = []

def agregar_paciente(nombre, cedula, genero, servicio):
    paciente = Paciente(nombre, cedula, genero, servicio)
    pacientes.append(paciente)

def agregar_enfermero(nombre, cedula, genero, turno, rango):
    enfermero = Enfermero(nombre, cedula, genero, turno, rango)
    enfermeros.append(enfermero)

def agregar_medico(nombre, cedula, genero, turno, especialidad):
    medico = Medico(nombre, cedula, genero, turno, especialidad)
    medicos.append(medico)

# Ejemplo de uso
agregar_paciente("Juan Perez", "123456789", "Masculino", "Cardiología")
agregar_enfermero("Maria Lopez", "987654321", "Femenino", "7-19", "Auxiliar")
agregar_medico("Dr. Rodriguez", "456789123", "Masculino", "19-7", "Cardiología")

# Imprimir información
print("Cédula de Pacientes:")
for paciente in pacientes:
    print(f"Nombre: {paciente._nombre}, Cédula: {paciente.ver_cedula()}")

print("\nCédula de Enfermeros:")
for enfermero in enfermeros:
    print(f"Nombre: {enfermero._nombre}, Cédula: {enfermero.ver_cedula()}")

print("\nCédula de Médicos:")
for medico in medicos:
    print(f"Nombre: {medico._nombre}, Cédula: {medico.ver_cedula()}")
```
