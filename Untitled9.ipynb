```python
class Pasajero:
    def __init__(self, nombre, apellido, edad, pasaporte):
        self._nombre = nombre
        self._apellido = apellido
        self._edad = edad
        self._pasaporte = pasaporte

    def mostrarInformación(self):
        return f"Pasajero {self._nombre} {self._apellido} \n con pasaporte {self._pasaporte}"


class Vuelo:
    def __init__(self, numero, origen, destino, capacidad):
        self.numero = numero
        self.origen = origen
        self.destino = destino
        self.capacidad = capacidad
        self.asientosDisponibles = capacidad
        self.reservas = []
    
    def mostrarInformación(self):
        return f"El Numero del vuelo es {self.numero} con origen {self.origen} \n Destino {self.destino} y su capacidad es {self.capacidad}"
    
    def reservaVuelo(self, pasajero):
        for p in self.reservas:
            if p._pasaporte == pasajero._pasaporte:
                return "El pasajero ya existe"
            
        if self.asientosDisponibles > 0:
            if pasajero._pasaporte.isdigit():
                self.reservas.append(pasajero)
                self.asientosDisponibles -= 1
                return f"Reserva exitosa para el pasajero {pasajero._nombre} {pasajero._apellido}"
            else:
                return "El pasaporte no es un número válido"
        else:
            return "Lo siento, no hay asientos disponibles en este vuelo"

    def verAsientosDisponibles(self):
        return f"El vuelo {self.numero} tiene {self.asientosDisponibles} asientos disponibles"

    # Esta es la función que agrega la funcionalidad de cancelar la reserva
    def cancelarReserva(self, pasaporte):
        for p in self.reservas:
            if p._pasaporte == pasaporte:
                self.reservas.remove(p)
                self.asientosDisponibles += 1
                return f"Reserva cancelada para el pasajero {p._nombre} {p._apellido}"
        
        return "No se encontró ningún pasajero con ese pasaporte"

class ReservaDeVuelos:
    # Este es el método constructor, que se ejecuta al crear un objeto de la clase
    def __init__(self):
        # Este es el atributo que almacena la lista de vuelos disponibles
        self.vuelos = []
    
    # Este es el método para mostrar la información de todos los vuelos disponibles
    def mostrarVuelos(self):
        # Se recorre la lista de vuelos y se llama al método mostrarInformación de cada uno
        for vuelo in self.vuelos:
            print(vuelo.mostrarInformación())
    
    # Este es el método para agregar un vuelo a la lista de vuelos disponibles
    def agregarVuelo(self, vuelo):
        # Se verifica que el vuelo sea un objeto de la clase Vuelo
        if isinstance(vuelo, Vuelo):
            # Se agrega el vuelo a la lista de vuelos
            self.vuelos.append(vuelo)
            # Se muestra un mensaje de confirmación
            print(f"Se ha agregado el vuelo {vuelo.numero} a la lista de vuelos disponibles")
        else:
            # Se muestra un mensaje de error
            print("El vuelo no es válido")

    # Esta es la función que agrega la funcionalidad de buscar vuelos según algunos criterios
    def buscarVuelos(self, origen=None, destino=None, capacidad=None):
        # Se crea una lista vacía para almacenar los vuelos que cumplan con los criterios
        resultados = []
        # Se recorre la lista de vuelos disponibles
        for vuelo in self.vuelos:
            # Se verifica si el origen, el destino y la capacidad coinciden con los del vuelo, o si son None (lo que significa que no se especificaron)
            if (origen == None or origen == vuelo.origen) and (destino == None or destino == vuelo.destino) and (capacidad == None or capacidad <= vuelo.capacidad):
                # Se agrega el vuelo a la lista de resultados
                resultados.append(vuelo)
        # Se verifica si se encontraron resultados
        if len(resultados) > 0:
            # Se muestra un mensaje con el número de resultados encontrados
            print(f"Se han encontrado {len(resultados)} vuelos que cumplen con los criterios")
            # Se muestra la información de cada resultado
            for resultado in resultados:
                print(resultado.mostrarInformación())
        else:
            # Se muestra un mensaje indicando que no se encontraron resultados
            print("No se han encontrado vuelos que cumplan con los criterios")

```

```python


# Se crea un objeto de tipo ReservaDeVuelos
reserva = ReservaDeVuelos()

# Se crean algunos objetos de tipo Vuelo con su información
vuelo = Vuelo("AA123", "Bogotá", "Medellín", 3)
vuelo1 = Vuelo(1234, "Bogotá", "Medellín", 42)
vuelo2 = Vuelo(5678, "Cali", "Cartagena", 36)
vuelo3 = Vuelo(9012, "Barranquilla", "Bucaramanga", 28)

# Se agregan los vuelos a la reserva
reserva.agregarVuelo(vuelo)
reserva.agregarVuelo(vuelo1)
reserva.agregarVuelo(vuelo2)
reserva.agregarVuelo(vuelo3)

pasajero1 = Pasajero("Juan", "Pérez", 30, "123456789")
pasajero2 = Pasajero("Juan", "Arroita", 30, "3744673")
pasajero3 = Pasajero("Juan", "Tolomeo", 30, "2372842")

vuelo = Vuelo("AA123", "Bogotá", "Medellín", 3)

print("\n=======VUELOS DISPONIBLES=========")


print(vuelo.mostrarInformación())
print(vuelo1.mostrarInformación())
print(vuelo2.mostrarInformación())
print(vuelo3.mostrarInformación())


print("\n==========BUSCAR VUELOS=========")
reserva.buscarVuelos(origen="Bogotá", destino="Medellín")
print("-------------------------------")
# Se busca un vuelo con capacidad mayor o igual a 40
reserva.buscarVuelos(capacidad=40)
print("-------------------------------")
# Se busca un vuelo sin especificar ningún criterio
reserva.buscarVuelos(destino="Medellín")

print("\n===========RESERVAS===========")
resultado1 = vuelo.reservaVuelo(pasajero1)
print(resultado1)
resultado2 = vuelo.reservaVuelo(pasajero2)
print(resultado2)
resultado3 = vuelo1.reservaVuelo(pasajero3)
print(resultado3)

print("\n==========RESERVAS CANCELADAS===========")
print(vuelo.cancelarReserva("123456789")) # Muestra "Reserva cancelada para el pasajero Juan Pérez"

print("\n===========ASIENTOS DISPONIBLES===========")

print(vuelo.verAsientosDisponibles())
print(vuelo1.verAsientosDisponibles())
print(vuelo2.verAsientosDisponibles())
print(vuelo3.verAsientosDisponibles())
```
